```python
from typing import Union
import pathlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import deeperhistreg
from deeperhistreg.dhr_input_output.dhr_loaders import pil_loader
from deeperhistreg.dhr_input_output.dhr_loaders import tiff_loader
from deeperhistreg.dhr_pipeline.registration_params import default_initial_nonrigid
```

```python
source_path = "HD_HE.jpg"
target_path = "CODEX_DAPI.jpg"

source_loader = pil_loader.PILLoader(source_path)
target_loader = pil_loader.PILLoader(target_path)

source = source_loader.load_level(level=0)
target = target_loader.load_level(level=0)

print("source:", source.shape)
print("target:", target.shape)
```

```python
plt.figure(dpi=200)
plt.imshow(source)
plt.axis('off')

plt.figure(dpi=200)
plt.imshow(target)
plt.axis('off')

source_patch = source_loader.load_region(level=0, offset=(900, 7000), shape=(1024, 1024))
target_patch = target_loader.load_region(level=0, offset=(900, 7000), shape=(1024, 1024))

plt.figure(dpi=200)
plt.imshow(source_patch)
plt.axis('off')

plt.figure(dpi=200)
plt.imshow(target_patch)
plt.axis('off')

plt.show()
```

```python
output_path = "."

source_path = pathlib.Path(source_path)
target_path = pathlib.Path(target_path)
output_path = pathlib.Path(output_path)

registration_params:dict = default_initial_nonrigid()
registration_params['loading_params']['loader'] = 'pil'
save_displacement_field:bool = True
copy_target:bool = True
delete_temporary_results:bool = False
case_name:str=f"{source_path.stem}-{target_path.stem}"
temporary_path:Union[str, pathlib.Path] = output_path/f"{pathlib.Path(source_path).stem}_{pathlib.Path(target_path).stem}_TEMP"

config = dict()
config['source_path'] = target_path
config['target_path'] = source_path
config['output_path'] = output_path
config['registration_parameters'] = registration_params
config['case_name'] = case_name
config['save_displacement_field'] = save_displacement_field
config['copy_target'] = copy_target
config['delete_temporary_results'] = delete_temporary_results
config['temporary_path'] = temporary_path
```

```python
deeperhistreg.run_registration(**config)
```

```python
warp_source_path = "warped_source.tiff"
target_path = "target.jpg"

warp_source_loader = pil_loader.PILLoader(warp_source_path)
target_loader = pil_loader.PILLoader(target_path)

warp_source = warp_source_loader.load_level(level=0)
target = target_loader.load_level(level=0)

print("warp source:", warp_source.shape)
print("target:", target.shape)
plt.figure(dpi=200)
plt.imshow(warp_source)
plt.axis('off')

plt.figure(dpi=200)
plt.imshow(target)
plt.axis('off')
```
